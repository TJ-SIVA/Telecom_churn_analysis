#Importing Libraries

```python
#importing essential libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#Loading Dataset

```python
#Loading raw csv file

df = pd.read_csv('Customer.csv')

#Converting CSV into feather file for reduce Space and Time Complexity

df.to_feather('customer.feather')
customer = pd.read_feather('customer.feather')
```

#Task 1 Understand the Dataset
###Description:
Familiarize yourself with the dataset.
####Steps:
* Load the dataset using pandas.

*  Display the first 10 rows.

* Identify the data types of each column.

* Check for missing values.

```python
# Step 1: Preview First Few Records

print("First 10 rows of the dataset:")
display(customer.head(10))

# Step 2: Dataset Structure and Data Types

print("\n\nDataset Info:\n")
customer.info()

# Step 3: Missing Value Check

print("\nMissing values per column:\n")
print(customer.isnull().sum())

#Visualize Missing Value Distribution (Only if nulls exist)

#Checking null value is exist...
if customer.isnull().sum().sum() > 0:
    sns.heatmap(customer.isnull(), cbar=False, cmap="YlOrRd")
    plt.title("Missing Value Heatmap")
    plt.show()
else:
    print("\nNo missing values — heatmap skipped.")

```
